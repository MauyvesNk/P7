### Sélection des variables catégorielles dépendantes

L'objectif ici est de sélectionner celles qui ont un lien statistique avec la variable cible.

Pour cela, nous allons utiliser le **Test du χ2 d'indépendance.**

Ce test permet de vérifier l'absence de lien statistique entre deux variables X et Y. Les deux sont dites indépendantes lorsqu'il n'existe aucun lien statistique entre elles, dit autrement, la connaissance de X ne permet en aucune manière de se prononcer sur Y. L'hypothèse nulle (H0) de ce test est la suivante : les deux variables X et Y sont indépendantes.

En termes de valeur p, l'hypothèse nulle est généralement rejetée lorsque p ≤ 0,05.

```python
# Importation de la fonction chi2_contingency de la bibliothèque scipy.stats
from scipy.stats import chi2_contingency
# Extraction de la variable cible 'target' du DataFrame X_train
y = X_train['target']
# Initialisation d'une liste pour stocker les valeurs de p-value résultant du test du chi2 pour chaque variable catégorielle
chi_ls = []
# Boucle à travers les variables catégorielles dans X_train
for feature in X_train.select_dtypes(include='object'):
    
    # Création d'une table de contingence
    c = pd.crosstab(y, X_train[feature])
    
    # Test du chi2
    p_value = chi2_contingency(c)[1]
    chi_ls.append(p_value)
```

```python
# Configuration du style seaborn avec une taille de figure spécifiée et une palette de couleurs 'BrBG'
sns.set(style="ticks", rc={'figure.figsize': (10, 10)}, palette="BrBG")
# Création d'un DataFrame contenant les valeurs de p-value résultant du test du chi2 pour chaque variable catégorielle
df_pvalue = pd.DataFrame(chi_ls, index=X_train.select_dtypes(include='object').columns, columns=['p_value']).sort_values(by='p_value', ascending=True)
# Configuration pour afficher les valeurs de p-value avec une précision de 3 chiffres après la virgule
pd.set_option('display.float_format', '{:,.3f}'.format)
# Affichage du DataFrame contenant les p-values
print(df_pvalue)
# Création d'un graphique à barres horizontales pour visualiser l'importance des variables basée sur le test du chi2
df_pvalue.plot.barh()
plt.xlabel('p value') # Libellé de l'axe des x
plt.title('Feature importance based on chi-square test') # Titre du graphique
plt.show()
```

## Les méthodes de filtrage de sélection des variables

### Constant features

Les caractéristiques constantes sont celles qui affichent la même valeur, une seule valeur, pour toutes les observations de l'ensemble de données. En d'autres termes, la même valeur pour toutes les lignes du jeu de données. Ces fonctionnalités ne fournissent aucune information permettant à un modèle d'apprentissage automatique de discriminer ou de prédire une cible.

L'identification et la suppression de fonctionnalités constantes constituent une première étape facile vers la sélection de fonctionnalités et des modèles d'apprentissage automatique plus facilement interprétables.

Pour identifier les caractéristiques constantes, nous pouvons utiliser le VarianceThreshold de Scikit-learn, ou nous pouvons le coder nous-mêmes. Si vous utilisez le VarianceThreshold, toutes nos variables doivent être numériques. Cependant, si nous le faisons manuellement, nous pouvons appliquer le code à la fois aux variables numériques et catégorielles.

Le VarianceThreshold de sklearn fournit une approche de base simple pour la sélection des fonctionnalités. Il supprime toutes les fonctionnalités dont la variance n'atteint pas un certain seuil. Par défaut, il supprime toutes les caractéristiques à variance nulle, c'est-à-dire les caractéristiques qui ont la même valeur dans tous les échantillons.

```python
# Importation de la classe VarianceThreshold du module sklearn.feature_selection
from sklearn.feature_selection import VarianceThreshold
# Création d'une instance de la classe VarianceThreshold avec un seuil de variance nul
selec = VarianceThreshold(threshold=0)
# Application de la méthode fit sur X_train pour identifier les caractéristiques avec une variance nulle
selec.fit(X_train)  # fit finds the features with zero variance
```

```python
# Calcul de la somme des caractéristiques sélectionnées par la méthode VarianceThreshold
sum(selec.get_support())
```

```python
# Affichage du nombre de caractéristiques constantes
# (utilisation de ~ pour exclure les caractéristiques non constantes)

# Sélection des colonnes constantes dans X_train à l'aide de selec.get_support()
constant = X_train.columns[~selec.get_support()]
# Calcul de la longueur (nombre) des caractéristiques constantes
len(constant)
```

Nous pouvons voir qu'il n'y a pas de colonne/variable numérique constante. Cela signifie qu'il n'y a pas de variable numérique qui affiche la même valeur, une seule valeur, pour toutes les observations de l'ensemble d'apprentissage.

Mais nous l'avions déjà vérifié dans la première partie du projet en analysant les données.

### Quasi-constant features

```python
# Création d'une liste vide pour stocker les caractéristiques quasi constantes
quasi_constant_feat = []

# Itération sur chaque caractéristique
for feature in X_train.columns:

    # Recherche de la valeur prédominante, c'est-à-dire la valeur partagée
    # par la plupart des observations
    predominant = X_train[feature].value_counts(
        normalize=True).sort_values(ascending=False).values[0]

    # Évaluation de la caractéristique prédominante : est-ce que plus de 99% des observations
    # montrent une seule valeur?
    if predominant > 0.998:

        # Si oui, ajout de la variable à la liste des caractéristiques quasi constantes
        quasi_constant_feat.append(feature)
# Calcul du nombre de caractéristiques quasi constantes
len(quasi_constant_feat)
```

Il n'y a pas de colonne/variable numérique et catégorielle constante. Cela signifie qu'il n'y a pas de variable qui affiche la même valeur, une seule valeur, pour toutes les observations de l'ensemble d'apprentissage.

Mais nous l'avions déjà vérifié dans la première partie du projet en analysant les données.

### Duplicated features

```python
# Vérification des caractéristiques dupliquées dans le jeu d'entraînement :

# Création d'un dictionnaire vide, où nous stockerons 
# les groupes de caractéristiques dupliquées
duplicated_feat_pairs = {}

# Création d'une liste vide pour collecter les caractéristiques
# qui ont été identifiées comme dupliquées
_duplicated_feat = []


# Itération sur chaque caractéristique de notre ensemble de données :
for i in range(0, len(X_train.columns)):
    
    # Ce morceau de code aide à comprendre où se trouve la boucle :
    if i % 10 == 0:  
        print(i)
    
    # Choix d'une caractéristique :
    feat_1 = X_train.columns[i]
    
   # Vérifier si cette caractéristique a déjà été identifiée
    # comme une dupliquée d'une autre. Si c'est le cas, elle devrait être stockée dans
    # notre liste _duplicated_feat.
    
    # Si cette caractéristique a déjà été identifiée comme une dupliquée, on la saute, si
    # elle n'a pas encore été identifiée comme une dupliquée, alors on procède :
    if feat_1 not in _duplicated_feat:
    
        # Création d'une liste vide comme entrée pour cette caractéristique dans le dictionnaire :
        duplicated_feat_pairs[feat_1] = []

        # Maintenant, itération sur les caractéristiques restantes de l'ensemble de données :
        for feat_2 in X_train.columns[i + 1:]:

            # Vérifier si cette deuxième caractéristique est identique à la première
            if X_train[feat_1].equals(X_train[feat_2]):

                # Si elle est identique, l'ajouter à la liste dans le dictionnaire
                duplicated_feat_pairs[feat_1].append(feat_2)
                
                # Et l'ajouter à notre liste de surveillance des variables dupliquées
                _duplicated_feat.append(feat_2)
```

```python
# Exploration de notre liste de caractéristiques dupliquées
len(_duplicated_feat)
```

```python
# Chaque ensemble contient un groupe de caractéristiques corrélées
selec.correlated_feature_sets_
```

Nous pouvons sélectionner la caractéristique avec la variance la plus élevée de chaque groupe.

```python
# Sélecteur de corrélation
sel = SmartCorrelatedSelection(
    variables=None,
    method="pearson",
    threshold=0.8,
    missing_values="raise",
    selection_method="variance",
    estimator=None,
    scoring="roc_auc",
    cv=3,
)
# Application du sélecteur sur X_train avec les étiquettes y
sel.fit(X_train, y)
```

```python
# Examinons la variance des caractéristiques du premier groupe de
# celles qui sont corrélées
group = list(sel.correlated_feature_sets_[0])
# Affichage de l'écart-type des caractéristiques dans X_train appartenant au groupe
X_train[group].std()
```

```python
# Examinons la variance des caractéristiques du deuxième groupe de
# celles qui sont corrélées
group =list(sel.correlated_feature_sets_[1])
# Affichage de l'écart-type des caractéristiques dans X_train appartenant au groupe
X_train[group].std()
```

```python
# Vérifier la longueur de correlated_feature_sets_
print(len(sel.correlated_feature_sets_))

# Si la longueur est suffisante, accéder à l'élément à l'indice 3
if len(sel.correlated_feature_sets_) > 3:
    group = sel.correlated_feature_sets_[3]
    # Affichage de l'écart-type des caractéristiques dans X_train appartenant au groupe
    X_train[group].std()
else:
    print("L'indice 3 est en dehors de la plage valide.")
```

```python
# Liste des caractéristiques à supprimer identifiées par le sélecteur
sel.features_to_drop_
```

```python
# Supprimer les caractéristiques corrélées
X_train = sel.transform(X_train)
X_test = sel.transform(X_test)
# Affichage des dimensions des ensembles de données après la suppression des caractéristiques corrélées
X_train.shape, X_test.shape
```

**Limites des méthodes de filtrage :**<br>

Les méthodes de filtrage sélectionnent les fonctionnalités indépendamment de tout classificateur spécifique. Par conséquent, ***l'un des inconvénients des méthodes de filtrage est qu'elles tendent à ignorer les effets des sous-ensembles de fonctionnalités sur les performances de l'algorithme de prédiction.***

De plus, les méthodes de filtrage évaluent souvent les variables individuellement et sont donc également **inconscientes du fait que certaines variables peuvent ne pas être prédictives prises isolément mais qu’elles le soient lorsqu'elles sont combinées avec d'autres variables**.

### Step Forward Feature Selection

```python
# Transformation des matrices en dataframe pour avoir les noms des variables
X_train_sd_df = pd.DataFrame(X_train_sd, columns=list(X_Xtrain.drop(labels='sk_id_curr', axis=1)))
X_train_sd_df.head()
```

```python
# Modèle de sélection forward
# Création d'une instance de SequentialFeatureSelector (SFS) avec un classificateur RandomForestClassifier
sfs1 = SFS(RandomForestClassifier(n_estimators=10, random_state=101), 
           k_features="best", 
           forward=True, 
           floating=False, 
           scoring='roc_auc',
           cv=5,
           n_jobs=-2
         )
```

```python
# Entraînement du modèle SequentialFeatureSelector (SFS) sur les données d'entraînement mises à l'échelle
sfs1 = sfs1.fit(X_train_sd_df, y_train_series)
# Affichage des sous-ensembles sélectionnés à chaque étape du processus de sélection
sfs1.subsets_
```

```python
# Configuration pour afficher toutes les colonnes du DataFrame
pd.set_option('display.max_colwidth', None)
# Création d'un DataFrame à partir des résultats du processus de sélection des caractéristiques
df = pd.DataFrame.from_dict(sfs1.get_metric_dict()).T
# Tri des résultats par score moyen décroissant
df.sort_values('avg_score', inplace=True, ascending=False)
# Affichage des premières lignes du DataFrame trié
df.head()
```

```python
# Importation de la fonction plot_sequential_feature_selection du module mlxtend.plotting
from mlxtend.plotting import plot_sequential_feature_selection as plot_sfs
# Création d'un graphique pour visualiser la sélection séquentielle des caractéristiques
fig1 = plot_sfs(sfs1.get_metric_dict(), kind='std_dev', figsize=(14, 6))
# Réglage des limites de l'axe y
plt.ylim([0.575, 0.650])
# Ajout de titre au graphique
plt.title('Sequential Forward Selection (w. StdDev)', fontsize=18, fontstyle='italic', fontweight='bold')
# Affichage du graphique
plt.show()
```

```python
# Liste des caractéristiques sélectionnées par le modèle SequentialFeatureSelector (SFS)
selected_feature_sfs1 = list(sfs1.k_feature_names_)
# Affichage du nombre de variables sélectionnées et de la liste des noms de variables
print(f"Nombre de variables sélectionnées : {len(selected_feature_sfs1)}")
print(selected_feature_sfs1)
```

### Recursive Feature Selection using Random Forests importance

```python
# Importation du modèle RandomForestClassifier, de la classe RFE (Recursive Feature Elimination) et de la métrique roc_auc_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import RFE
from sklearn.metrics import roc_auc_score
```

```python
# Création d'une instance de RFE avec RandomForestClassifier comme modèle et ajustement sur les données d'entraînement
sel_ = RFE(RandomForestClassifier(n_estimators=10, random_state=101))
sel_.fit(X_train_sd, y_train)
```

## Comparaison des performances des 2 sous-ensembles de fonctionnalités

```python
def run_randomForests(X_train, X_test, y_train, y_test):
     # Initialisation du modèle Random Forest
    rf = RandomForestClassifier(n_estimators=200, random_state=101, max_depth=4)
    # Entraînement sur l'ensemble d'entraînement
    rf.fit(X_train, y_train)
    # Évaluation sur l'ensemble d'entraînement
    print('Train set')
    pred = rf.predict_proba(X_train)
    print('Random Forests roc-auc: {}'.format(roc_auc_score(y_train, pred[:,1])))
    # Évaluation sur l'ensemble de test
    print('Test set')
    pred = rf.predict_proba(X_test)
    print('Random Forests roc-auc: {}'.format(roc_auc_score(y_test, pred[:,1])))
```

```python
# Création de DataFrames pour les données de validation après mise à l'échelle
X_valid_sd_df = pd.DataFrame(X_valid_sd, columns=list(X_valid.drop(labels='sk_id_curr', axis=1)))
# Conversion de la série y de validation en DataFrame
y_valid_series = pd.Series(y_valid)
```

```python
# Exécution de Random Forests avec les fonctionnalités sélectionnées récursivement
run_randomForests(X_train_sd_df[selected_feat_rfe],
                  X_valid_sd_df[selected_feat_rfe],
                  y_train_series, y_valid_series)
```

```python
# Exécution de Random Forests avec toutes les fonctionnalités sélectionnées en même temps
run_randomForests(X_train_sd_df[selected_feat],
                  X_valid_sd_df[selected_feat],
                  y_train_series, y_valid_series)
```

Nous voyons que RFE a renvoyé un sous-ensemble de fonctionnalités légèrement meilleur que le randomforest simple.
Nous sélectionnons donc l'ensemble des variables issu de la méthode *Recursive Feature Selection* (RFE)

```python
# Sélection des fonctionnalités pour le modèle avec les caractéristiques sélectionnées de manière récursive
X_train_clean = X_train[selected_feat_rfe]
X_test_clean = X_test[selected_feat_rfe]
# Affichage des dimensions des ensembles de données après la sélection des fonctionnalités
X_train_clean.shape, X_test_clean.shape
```
